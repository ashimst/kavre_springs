```python
def clean_measures(x):
    if pd.isna(x):
        return []
    x = x.lower().strip()
    x = x.replace('source protection', 'source_protection')  # unify naming
    if x == '0' or x == 'tried':
        return []
    return [item.strip() for item in x.split(',')]

df['measures_taken_clean'] = df['measures_taken'].apply(clean_measures)

mlb = MultiLabelBinarizer()
one_hot_encoded = pd.DataFrame(mlb.fit_transform(df['measures_taken_clean']),
                               columns=mlb.classes_,
                               index=df.index)

# Before concatenating, drop columns if they already exist
for col in one_hot_encoded.columns:
    if col in df.columns:
        df.drop(columns=[col], inplace=True)

df = pd.concat([df, one_hot_encoded], axis=1)

print(df.columns.tolist())
```

```python
def clean_roles(text):
    if pd.isna(text):
        return ''
    text = str(text).lower().strip()
    text = text.replace('&', ',')
    text = text.replace('treserurer', 'treasurer')
    text = text.replace('teserurer', 'treasurer')
    text = text.replace('secretrary', 'secretary')
    text = text.replace('vice-chair', 'vice_chair')
    if text in ['0', '', ' ']:
        return ''
    return text

df['Women in leadership position'] = df['Women in leadership position'].apply(clean_roles)

df['leadership_list'] = df['Women in leadership position'].apply(
    lambda x: [role.strip() for role in x.split(',')] if x else []
)

mlb = MultiLabelBinarizer()
one_hot = pd.DataFrame(mlb.fit_transform(df['leadership_list']),
                       columns=mlb.classes_,
                       index=df.index)

one_hot = one_hot.add_prefix('f_')

df = df.join(one_hot)
df.drop(columns=['leadership_list','Women in leadership position'], inplace=True)
```

```python
def map_to_category(text):
    if pd.isna(text):
        return ['misc_unknown']
    text = text.lower()
    categories = []

    if any(k in text for k in ['bihe', 'bibaha', 'बिहे']):
        categories.append('marriage')
    if any(k in text for k in ['brata', 'वरतमान', 'bratabandha']):
        categories.append('bratabandha')
    if any(k in text for k in ['kriya', 'किरिया', 'pitri', 'श्राद्ध']):
        categories.append('kriya_funeral')
    if any(k in text for k in ['nag', 'नाग']):
        categories.append('nag_puja')
    if any(k in text for k in ['panchami', 'पञ्चमि']):
        categories.append('panchami')
    if any(k in text for k in ['teej', 'तिज']):
        categories.append('teej')
    if any(k in text for k in ['mandir', 'मन्दिर']):
        categories.append('mandir_related')
    if any(k in text for k in ['nity', 'नित्य', 'niyamit', 'दैनिक', 'घर']):
        categories.append('regular_worship')
    if any(k in text for k in ['shradh', 'पिण्ड', 'स्राद्ध', 'पित्रि']):
        categories.append('shraddha')
    if any(k in text for k in ['dashain', 'sivaratri', 'mela', 'holi', 'maghe']):
        categories.append('other_festivals')

    if not categories:
        return ['misc_unknown']
    return categories

# Apply the function
df['event_categories'] = df['List the events associated with sources'].apply(map_to_category)
```

```python
mlb = MultiLabelBinarizer()
encoded = mlb.fit_transform(df['event_categories'])

category_df = pd.DataFrame(encoded, columns=mlb.classes_)
df = pd.concat([df, category_df], axis=1)
```
